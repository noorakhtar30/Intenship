```python
#variable
a = 0.5
b = 1
c = 2
d = 3
```

# Finding the best goal keeper 

```python
df['gk_Shotstopper'] = (b*df.Reactions + c*df.Jumping + b*df.GK_Positioning + d*df. GK_Reflexes +
                        b*df.GK_Handling)/(2*a+ 4*b + 2*c + 1*d)
df['gk_Sweeper'] = (b*df.Reactions+ b*df.Jumping +c*df.Speed + a*df.Stamina + b*df.GK_Positioning +
                    b*df.GK_Diving + d*df. GK_Reflexes + b*df.GK_Reflexes + b*df.GK_Handling +
                    d*df.GK_Kicking + c*df.Vision)/(2*a+ 4*b + 2*c + 1*d)
```

```python
plt.figure(figsize=(15,6))
sd = df.sort_values('gk_Shotstopper',ascending=False)[:5]
x = np.array(list(sd['Name']))
y = np.array(list(sd['gk_Shotstopper']))
sns.barplot(x, y, palette= sns.color_palette("gist_yarg_r"))
plt.ylabel('Shot Stopping Score')
```

```python
plt.figure(figsize=(15,6))
sd = df.sort_values('gk_Sweeper',ascending=False)[:5]
x1 = np.array(list(sd['Name']))
y1 = np.array(list(sd['gk_Sweeper']))
sns.barplot(x1, y1, palette= sns.color_palette("viridis_r"))
plt.ylabel('Sweeping Score')
```

# Finding the best Defenders

```python
df['def_centerbacks'] = ( d*df.Reactions + c*df.Interceptions + d*df.Sliding_Tackle + d*df.Standing_Tackle + b*df.Vision +
                         a*df.Short_Pass + b*df.Long_Pass + c*df.Acceleration + b*df.Speed + d*df.Stamina + d*df.Jumping + 
                         d*df.Heading + b*df.Long_Shots + d*df.Marking + c*df.Aggression)/(6*b + 3*c + 7*d)
```
