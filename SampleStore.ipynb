```python
# Importing the essential librearies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the csv file
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.info()
```

```python
# Checking the null values if any
df.isnull().sum()
```

### Countplot


```python
sns.countplot(x='Category',data=df)
```

```python
sns.countplot(x='Region',data=df)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=70)
```

### Heatmap

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True)
```

### Barplot

```python
plt.figure(figsize=(15,9))
sns.barplot(df['Sub-Category'],y=df['Profit'])
```

## 20 states with maxium sales done

```python
plt.figure(figsize=(15,8))
y=df.sort_values(['Sales'],ascending=False).head(20)
plt.bar(y['State'],y['Sales'])
plt.xlabel('States')
plt.ylabel('Sales')
plt.show()
```

## product used most in which region

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Category',hue='Region',data=df)
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(20,20))
sns.boxplot(x='Sub-Category',y='Profit',data=df)
plt.show()
```
